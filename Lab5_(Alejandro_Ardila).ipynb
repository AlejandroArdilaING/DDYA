## 9) Código funcional



```python
pacientes_criticos = []
pacientes_normales = []

def main():

    cantidad = int(input("Cuantos pacientes va a ingresar: "))

    i = 0
    while i < cantidad:

        dato = input("Escriba el paciente (cNombre o nNombre): ")

        if dato[0] == "c":
            nombre = dato[1:]
            pacientes_criticos.append(nombre)

        elif dato[0] == "n":
            nombre = dato[1:]
            pacientes_normales.append(nombre)

        i = i + 1

    print("Criticos:", pacientes_criticos)
    print("Normales:", pacientes_normales)

    while len(pacientes_criticos) > 0 or len(pacientes_normales) > 0:

        if len(pacientes_criticos) > 0:
            print("Atendiendo paciente critico:", pacientes_criticos[0])
            pacientes_criticos.pop(0)

        elif len(pacientes_normales) > 0:
            print("Atendiendo paciente normal:", pacientes_normales[0])
            pacientes_normales.pop(0)

    print("No quedan pacientes")


main()
```
